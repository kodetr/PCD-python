```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
citra = cv2.imread('./Images/manggis1.jpg')
```

```python
plt.imshow(citra[:,:,[2,1,0]])
plt.show
```

```python
#Citra abu abu
gray_image = cv2.cvtColor(citra, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('Asli')
plt.imshow(citra[:,:,[2,1,0]])

plt.subplot(122)
plt.title('GrayScile')
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.show
```

```python
#[enghilang nois]
blurred_image = cv2.GaussianBlur(gray_image, (7, 7), 0)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('GrayScile')
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Smoot')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.show
```

```python
#deteksi tepi dengan menggunakan algoritma Sobel
# Terapkan deteksi tepi dengan operator Sobel
edges_x = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
edges_y = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)
edges = cv2.addWeighted(cv2.convertScaleAbs(edges_x), 0.5, cv2.convertScaleAbs(edges_y), 0.5, 0)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('Smoot')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Deteksi Tepi')
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.show
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Images/manggis.jpg')
```

```python
plt.imshow(img[:,:,[2,1,0]])
plt.show
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('Asli')
plt.imshow(img[:,:,[2,1,0]])

plt.subplot(122)
plt.title('GrayScale')
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.show
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
contrasted_image = clahe.apply(gray_image)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('GrayScale')
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Peninggkatan Kontras')
plt.imshow(cv2.cvtColor(contrasted_image, cv2.COLOR_BGR2RGB))
plt.show
```

```python
#citra biner
threshold_value = 225  # Ganti nilai ambang sesuai kebutuhan

#jika ingin objek berwarna hitam
#_, binary_image = cv2.threshold(contrasted_image, threshold_value, 255, cv2.THRESH_BINARY)

#jika ingin objek berwarna putih
_, binary_image = cv2.threshold(contrasted_image, threshold_value, 255, cv2.THRESH_BINARY_INV)

```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('Peningkatan Kontras')
plt.imshow(cv2.cvtColor(contrasted_image, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Citra Biner')
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))
plt.show
```

```python
denoised_image = cv2.medianBlur(binary_image, 5) 
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('Citra Biner')
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.title('Penghilang Nois')
plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
plt.show
```

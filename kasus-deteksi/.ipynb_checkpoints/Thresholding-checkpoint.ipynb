```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Images/kai.jpg')
```

```python
print(img)
```

```python
img.shape
```

```python
#Grayscale

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap ='gray')
plt.show
```

```python
threshold = 100
```

```python
#Thresholding

for y in range(img_gray.shape[0]): #tinggi (y)
    for x in range(img_gray.shape[1]): #lebar (x)
        if(img_gray[y,x] > threshold):
            img_gray[y,x] = 255
        else: 
            img_gray[y,x] = 0
```

```python
plt.imshow(img_gray, cmap ='gray')
plt.show()
```

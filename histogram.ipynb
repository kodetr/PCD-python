```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./images/messi.jpg')
```

```python
img.shape # tinggi, lebar -> (y,x)
```

```python
img.shape[0] # y
```

```python
img.shape[1] # x
```

# Display

```python
# BGR -> 012
# RGB -> 210
plt.imshow(img[:,:,[2,1,0]])
plt.show()
```

# Histogram

```python
hitogram_r = [0] * 256
hitogram_g = [0] * 256
hitogram_b = [0] * 256

for y in range(img.shape[0]): # 262
    for x in range(img.shape[1]): # 193
        
        intensitas_r = img[y,x,2]
        intensitas_g = img[y,x,1]
        intensitas_b = img[y,x,0]
        
        hitogram_r[intensitas_r] +=1
        hitogram_g[intensitas_g] +=1
        hitogram_b[intensitas_b] +=1
  
    
plt.figure(figsize = (8,5))
plt.subplot(311)
plt.bar(range(256), hitogram_r, width = 1, color = 'red')
plt.xlabel('Intensitas Merah')
plt.ylabel('Jumlah piksel')
plt.title('Histogram Channel Merah')

plt.subplot(312)
plt.bar(range(256), hitogram_g, width = 1, color = 'green')
plt.xlabel('Intensitas Hijau')
plt.ylabel('Jumlah piksel')
plt.title('Histogram Channel Hijau')

plt.subplot(313)
plt.bar(range(256), hitogram_b, width = 1, color = 'blue')
plt.xlabel('Intensitas Biru')
plt.ylabel('Jumlah piksel')
plt.title('Histogram Channel Biru')

plt.tight_layout()
plt.show()
        
```

# Hitogram Grayscale

```python
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('MESSI', grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
grayscale.shape
```

```python
hitogram = [0] * 256

for y in range(grayscale.shape[0]): # 262
    for x in range(grayscale.shape[1]): # 193
        
        intensitas = grayscale[y,x]
        
        hitogram[intensitas] +=1
     
    
plt.figure(figsize = (8,5))
plt.subplot(311)
plt.bar(range(256), hitogram, width = 1, color = 'black')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah')
plt.title('Histogram')       

plt.tight_layout()
plt.show()
```

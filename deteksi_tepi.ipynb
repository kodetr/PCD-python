# Conversi BGR to GRAYSCALE

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
plt.show()
```

# Deteksi Tepi

```python
deteksi_tepi = np.zeros_like(img_gray)

threshold = 50
```

```python
for y in range(img_gray.shape[0] - 1):
    for x in range(img_gray.shape[1] -1):
        dx = int(img_gray[y,x+1]) - int(img_gray[y,x-1])
        dy = int(img_gray[y+1,x]) - int(img_gray[y-1,x])
     
        total_gradien = np.sqrt(dx**2 + dy**2)
        
        if(total_gradien > threshold):
            deteksi_tepi[y,x] = 255
```

```python
plt.imshow(deteksi_tepi, cmap='gray')
plt.show()
```

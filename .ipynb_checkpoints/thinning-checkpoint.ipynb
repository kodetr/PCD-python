```python
import cv2
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
```

```python
img = cv2.imread('images/gambar1.bmp', 0)
```

```python
img
```

# Display

```python
plt.imshow(img, cmap='gray')
plt.show()
```

# Conversi biner dan thresholding

```python
_, binner_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binner_img, cmap='gray')
plt.show()
```

# Thinning

```python
thinning = skeletonize(binner_img // 255)
```

```python
plt.imshow(thinning, cmap='gray')
plt.show()
```

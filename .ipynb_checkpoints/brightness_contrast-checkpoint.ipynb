```python
import cv2
import numpy as np # clip -> untuk memastikan bahwa nilai piksel tetap dalam range 0 -> 255
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./images/messi.jpg') # gambar 8 bit BGR
```

```python
print(img)
```

# Brightness / Kecerahan

Piksel Baru = Piksel Lama + Brighness

```python
brightness = 150 # range -255 (gelap) <-> 0 (normal) <-> 255


image_brightness = np.clip(img.astype(int) + brightness, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img[:,:,[2,1,0]]) # hasil original
plt.title('Original')

plt.subplot(122)
plt.imshow(image_brightness[:,:,[2,1,0]]) # hasil brightness
plt.title('Brightness')
plt.show()
```

# Contrast

Piksel Baru = (Piksel Lama - 128) x contrast + 128

```python
contrast = -2.5 # range -128 <-> 0 (normal) <-> 128

image_contrast = np.clip((img.astype(float) - 128) * contrast + 128, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img[:,:,[2,1,0]]) # hasil original
plt.title('Original')

plt.subplot(122)
plt.imshow(image_contrast[:,:,[2,1,0]]) # hasil brightness
plt.title('Brightness')
plt.show()
```
